```

import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
Wf, Uf, bf = 0.5, 0.1, 0
Wi, Ui, bi = 0.6, 0.2, 0
Wc, Uc, bc = 0.7, 0.3, 0
Wo, Uo, bo = 0.8, 0.4, 0

Wy = 4
by = 0


X = [1, 2, 3, 4]

h = 0
C = 0

print("LSTM Numerical Example\n")

for t, x in enumerate(X, start=1):

    print(f"Time Step {t}  |  Input = {x}")

    f = sigmoid(Wf*x + Uf*h + bf)
    print("Forget gate =", round(f,3))

    i = sigmoid(Wi*x + Ui*h + bi)
    print("Input gate =", round(i,3))

    C_tilde = tanh(Wc*x + Uc*h + bc)
    print("Candidate cell =", round(C_tilde,3))

    C = f*C + i*C_tilde
    print("Cell state =", round(C,3))

    o = sigmoid(Wo*x + Uo*h + bo)
    print("Output gate =", round(o,3))

    h = o * tanh(C)
    print("Hidden state =", round(h,3))

    print()

y_hat = Wy*h + by

print("Final Prediction =", round(y_hat,3))
```
